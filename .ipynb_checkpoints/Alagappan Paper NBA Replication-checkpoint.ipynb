```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Reading the data

```python
df = (
    pd.read_csv("data/nba_2010-2011_per_game.csv") #, index_col="PLAYERS"
    .query("Tm != 'TOT'")
    .drop_duplicates(subset='Player', keep='first') # First entry will always be Tm=TOT
    .set_index("Player")
    .sample(frac=1.0)
)
df['Pos'] = df['Pos'].str.split('-').str[0]
display(df.head())
```

```python
print(f"Number of unique players: {df.index.nunique()}")
print(f"Number of rows in the dataset: {df.shape[0]}")
```

# Getting the features

```python
features = ['PTS', 'AST', 'TOV', 'TRB', 'STL', 'BLK','3P', 'FTA', 
            'Pos', 'Tm'] #, 'MP'
df_features = df[features].copy() #.reset_index(drop=True)
position_encoding = pd.get_dummies(df_features["Pos"])[['PG', 'SG', 'SF', 'PF','C']]
team_encoding = pd.get_dummies(df_features["Tm"])
df_features_no_pos_tm = df_features.drop(["Pos", "Tm"], axis=1)
df_features.head()
```

# DBSCAN

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import DBSCAN

def get_kdist(k, data):
    # Recall that MinPts = twice the number of dimensions and k = MinPts - 1
    # We compute the distance of each point from its (k+1)th-nearest neighbor
    neigh = NearestNeighbors(n_neighbors=k+1) # we use k+1 to disregard the point itself
    nbrs = neigh.fit(data)
    distances, _ = nbrs.kneighbors(data)
    return list(np.sort(distances[:, k])[::-1])

# MinMax Scaling
min_pts = df_features.drop('Pos', axis=1).shape[1] * 2
k = min_pts - 1

plt.plot(get_kdist(k, df_features_no_pos_tm))
plt.xlabel("point")
plt.ylabel("distance");
```

# Giotto-TDA

### Defining the mapper pipeline

```python
from gtda.mapper import (
    CubicalCover,
    make_mapper_pipeline,
    Projection,
    plot_static_mapper_graph,
    plot_interactive_mapper_graph,
    MapperInteractivePlotter
)
from sklearn.decomposition import PCA # dimensionality reduction
from umap import UMAP

# UMAP lens
umap_lens = UMAP(n_components=2, random_state=42) # UMAP

# Initialise pipeline
mapper_pipeline = make_mapper_pipeline(
    filter_func=umap_lens, #PCA(n_components=2), 
    cover=CubicalCover(n_intervals=22, overlap_frac=0.3), # for umap
    clusterer=DBSCAN(), #eps=4.5, min_samples=min_pts
    verbose=True,
    n_jobs=-1
)
```

### Helper functions for coloring the nodes by Stats, Position, and Team

```python
# Color by stats 
def stats_plotter():
    fig = plot_static_mapper_graph(mapper_pipeline, 
                               df_features_no_pos_tm, 
                                   #layout="fruchterman_reingold",
                               layout_dim=2, 
                               color_data=df_features_no_pos_tm)
    fig.update_layout(
        #paper_bgcolor='lightgray',  # This sets the color outside the plot area
        #plot_bgcolor='lightgray',    # This sets the color of the plot area
        width=700, height=500
    )
    fig.show(config={'scrollZoom': False})


# Color by positions 
def pos_plotter():
    fig = plot_static_mapper_graph(mapper_pipeline, 
                               df_features_no_pos_tm,
                                   #layout="fruchterman_reingold",
                               layout_dim=2, 
                               color_data=position_encoding)
    fig.update_layout(
        #paper_bgcolor='lightgray',  # This sets the color outside the plot area
        #plot_bgcolor='lightgray',    # This sets the color of the plot area
        width=700, height=500
    )
    fig.show(config={'scrollZoom': False})


# Color by Team 
def team_plotter():
    fig = plot_static_mapper_graph(mapper_pipeline, 
                               df_features_no_pos_tm,
                                #layout="fruchterman_reingold",
                               layout_dim=2, 
                               color_data=team_encoding)
    fig.update_layout(
        #paper_bgcolor='lightgray',  # This sets the color outside the plot area
        #plot_bgcolor='lightgray',    # This sets the color of the plot area
        width=700, height=500
    )
    fig.update_layout()
    fig.show(config={'scrollZoom': False})
```

# Plots
